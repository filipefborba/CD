```python
dados = pd.read_table("pes2013A.txt", encoding="windows-1252", sep=" ")
dados.head()
```

```python
dados.UF = dados.UF.astype("category")
dados.UF.cat.categories = ("Rondônia",\
"Acre",\
"Amazonas",\
"Roraima",\
"Pará",\
"Amapá",\
"Tocantins",\
"Maranhão",\
"Piauí",\
"Ceará",\
"Rio Grande do Norte",\
"Paraíba",\
"Pernambuco",\
"Alagoas",\
"Sergipe",\
"Bahia",\
"Minas Gerais",\
"Espírito Santo",\
"Rio de Janeiro",\
"São Paulo",\
"Paraná",\
"Santa Catarina",\
"Rio Grande do Sul",\
"Mato Grosso do Sul",\
"Mato Grosso",\
"Goiás",\
"Distrito Federal")
dados.V0302 = dados.V0302.astype("category")
dados.V0302.cat.categories = ("Masculino","Feminino")
dados.V0404 = dados.V0404.astype("category")
dados.V0404.cat.categories = ("Indígena","Branca","Preta","Amarela","Parda","Sem Declaração")
dados.V6002 = dados.V6002.astype("category")
dados.V6002.cat.categories = ("Pública","Privada")
dados.V4802 = dados.V4802.astype("category")
dados.V4802.cat.categories = ("Maternal, jardim de infância etc.", "Classe de alfabetização", "Alfabetização de adultos",\
"1ª a 4ª séries do ensino fundamental regular em 8 anos ou de educação de jovens e adultos",\
"5ª a 8ª séries do ensino fundamental regular em 8 anos ou de educação de jovens e adultos",\
"1ª série do ensino fundamental regular em 9 anos",\
"2ª a 5ª séries do ensino fundamental regular em 9 anos",\
"6ª a 9ª séries do ensino fundamental regular em 9 anos",\
"Educação de jovens e adultos não seriado do ensino fundamental",\
"Ensino médio, inclusive pré-vestibular",\
"Superior, inclusive mestrado ou doutorado")

slicedtable = dados.loc[:, ("UF","V0302","V0404","V6002","V4802")]
```

## Tabelas Univariadas

```python
#Tabelas univariadas
n=dados.shape[0] #Número de linhas

#Estado (UF)
print("Estado a que pertence")
table0 = (dados.UF.value_counts()/n*100).round(decimals=1).reindex(["Rondônia",\
"Acre",\
"Amazonas",\
"Roraima",\
"Pará",\
"Amapá",\
"Tocantins",\
"Maranhão",\
"Piauí",\
"Ceará",\
"Rio Grande do Norte",\
"Paraíba",\
"Pernambuco",\
"Alagoas",\
"Sergipe",\
"Bahia",\
"Minas Gerais",\
"Espírito Santo",\
"Rio de Janeiro",\
"São Paulo",\
"Paraná",\
"Santa Catarina",\
"Rio Grande do Sul",\
"Mato Grosso do Sul",\
"Mato Grosso",\
"Goiás",\
"Distrito Federal"])
print(table0,'\n')

#Sexo (V0302)
print('Porcentagem de Pessoas por Sexo:')
table1 = (dados.V0302.value_counts()/n*100).round(decimals=1).reindex(["Masculino", "Feminino"])
print(table1,'\n')

#Raça (V0404)
print('Porcentagem de Pessoas por Raça:')
table2 = (dados.V0404.value_counts()/n*100).round(decimals=1).reindex(["Indígena","Branca","Preta","Amarela","Parda","Sem Declaração"])
print(table2,'\n')

#Rede de ensino (V6002)
print('Porcentagem de Pessoas por Rede de Ensino:')
table3 = (dados.V6002.value_counts()/n*100).round(decimals=1).reindex(["Pública", "Privada","Não Aplicável"])
print(table3,'\n')

#Nível de ensino e grupos de séries do ensino fundamental que frequentavam (Todos os estudantes - V4802)
print('Porcentagem de Pessoas por Raça:')
table4 = (dados.V4802.value_counts()/n*100).round(decimals=1).reindex(["Maternal, jardim de infância etc.",\
"Classe de alfabetização",\
"Alfabetização de adultos",\
"1ª a 4ª séries do ensino fundamental regular em 8 anos ou de educação de jovens e adultos",\
"5ª a 8ª séries do ensino fundamental regular em 8 anos ou de educação de jovens e adultos",\
"1ª série do ensino fundamental regular em 9 anos",\
"2ª a 5ª séries do ensino fundamental regular em 9 anos",\
"6ª a 9ª séries do ensino fundamental regular em 9 anos",\
"Educação de jovens e adultos não seriado do ensino fundamental",\
"Ensino médio, inclusive pré-vestibular",\
"Superior, inclusive mestrado ou doutorado"])
print(table4,'\n')
```

## Tabelas cruzadas entre variáveis qualitativas

```python
#Função que divide cada valor da tabela cruzada pelo total da última linha (quando axis=0) ou da última coluna (quando axis=1)
#Retorna tabela cruzada com valores sem casa decimais
#https://www.analyticsvidhya.com/blog/2016/01/12-pandas-techniques-python-data-manipulation/
def percConvert(tab):
  return (tab/float(tab[-1])*100).round(decimals=0)
```

```python
#Tabela cruzada Estados x Raça
print('Tabela Cruzada, Estados x Raça, TOTAL EM UNIDADES:')
ct=pd.crosstab(dados.UF,dados.V0404,margins=True,rownames=['ESTADO'], colnames=['RAÇA'])
ct=ct.loc[["Rondônia",\
"Acre",\
"Amazonas",\
"Roraima",\
"Pará",\
"Amapá",\
"Tocantins",\
"Maranhão",\
"Piauí",\
"Ceará",\
"Rio Grande do Norte",\
"Paraíba",\
"Pernambuco",\
"Alagoas",\
"Sergipe",\
"Bahia",\
"Minas Gerais",\
"Espírito Santo",\
"Rio de Janeiro",\
"São Paulo",\
"Paraná",\
"Santa Catarina",\
"Rio Grande do Sul",\
"Mato Grosso do Sul",\
"Mato Grosso",\
"Goiás",\
"Distrito Federal",\
"All"]]
ct
```

```python
print('Tabela Cruzada, Estado x Raça: TOTAL EM %')
ct1=(ct/len(dados)*100).round(decimals=2)
ct1
```

```python
#Tabela cruzada dos estados pelos niveis de ensino
UFN = pd.crosstab(dados.UF, dados.V4802, margins=True,rownames=['ESTADO'], colnames=['NIVEL DE ENSINO'])
slicedUFN = UFN.iloc[:,(9,10)]
slicedUFN
```

```python
UFN1 = (UFN/len(dados)*100).round(decimals=2)
UFN1 = UFN1.iloc[:,(9,10,11)]
UFN1
```

```python
UFN1 = (UFN/len(dados)*100).round(decimals=2)
UFN1 = UFN1.iloc[0:27,(9,10)]
plot = UFN1.plot(kind='bar', stacked=True,title='Plano em cada Satisfação')
plt.ylabel('Freq.Relativa (em %)')
plot
```
