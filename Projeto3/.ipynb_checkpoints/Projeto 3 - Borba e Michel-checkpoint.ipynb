# Projeto 3 - Ciência dos Dados
#### Variável: Expectativa de Vida

Problemática: Como algumas necessidades básicas (Comida, Saneamento Básico, Educação) podem afetar a expectativa de vida no mundo?

```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import pandas as pd
import numpy as np

import statsmodels.api as sm1 #Usa OLS com X como matriz de planejamento
import statsmodels.formula.api as sm  #Usa OLS como no R

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = dados = pd.read_excel('Dados.xlsx',index_col=0)
```

```python
dados=dados[~np.isnan(dados.Life)]
dados=dados[~np.isnan(dados.Child)]
dados=dados[~np.isnan(dados.Sanitation)]
dados=dados[~np.isnan(dados.School)]
dados=dados[~np.isnan(dados.Food)]
print(dados.shape)
dados.head()
```

### Análise Descritiva

```python
f, ax = plt.subplots(2,2,figsize=(12,12))
ax[0,0].scatter(dados.Child, dados.Life, alpha=0.7)  # Plot the raw data
ax[0,0].set_xlabel("Taxa de Mortalidade Infantil (mortes por 1000 nascimentos)")
ax[0,0].set_ylabel("Expectativa de vida (em anos)")

ax[0,1].scatter(dados.Sanitation, dados.Life, alpha=0.7)  # Plot the raw data
ax[0,1].set_xlabel("% da população com Saneamento Básico")
ax[0,1].set_ylabel("Expectativa de vida (em anos)")

ax[1,0].scatter(dados.School, dados.Life, alpha=0.7)  # Plot the raw data
ax[1,0].set_xlabel("% de conclusão do Ensino Básico")
ax[1,0].set_ylabel("Expectativa de vida (em anos)")

ax[1,1].scatter(dados.Food, dados.Life, alpha=0.7)  # Plot the raw data
ax[1,1].set_xlabel("Ingestão de calorias (kcal/pessoa*dia)")
ax[1,1].set_ylabel("Expectativa de vida (em anos)")

plt.tight_layout(1)
```

```python
dados.corr(method='pearson')
```

```python
#Linear Regression - PRIMEIRA MANEIRA DE FAZER REGRESSÃO USANDO PACOTE ols
#Escreve modelo como no R

model = sm.ols(formula='Life ~ Sanitation + Child', data = dados)
fit = model.fit()
fit.summary()

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

x_surf = np.arange(0, 100, 1)                # generate a mesh
y_surf = np.arange(0, 200, 1)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

exog = pd.core.frame.DataFrame({'Sanitation': x_surf.ravel(), 'Child': y_surf.ravel()})
out = fit.predict(exog = exog)
ax.plot_surface(x_surf, y_surf,
                out.reshape(x_surf.shape),
                rstride=1,
                cstride=1,
                color='None',
                alpha = 0.1)

ax.scatter(dados['Sanitation'], dados['Child'], dados['Life'],
           c='blue',
           marker='o',
           alpha=0.7)

ax.set_xlabel('% da população com Saneamento Básico')
ax.set_ylabel('Taxa de Mortalidade Infantil (mortes por 1000 nascimentos)')
ax.set_zlabel('Expectativa de Vida (em anos)')
```

```python
#Linear Regression - SEGUNDA MANEIRA DE FAZER REGRESSÃO USANDO PACOTE ols
#Informando X como matriz de planejamento

#Montagem dos dados
y=dados.Life
X=dados.iloc[:,(1,2,3,4)]
X=sm1.add_constant(X)  # Adds a constant term to the predictor
X.head()

model=sm1.OLS(y, X)
fit=model.fit()
fit.summary()
```

```python
#Teste de White
#ver link https://pt.wikipedia.org/wiki/Teste_de_White
#y estimados
y_hat = fit.predict(X)

#Resíduos quadráticos
res2=(y-y_hat)**2

sm1.stats.diagnostic.het_white(res2,X)
```
