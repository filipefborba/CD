# Projeto 3 - Ciência dos Dados
#### Variável: Expectativa de Vida

Problemática: Como algumas necessidades básicas (Comida, Saneamento Básico, Educação) podem afetar a expectativa de vida no mundo?

#### Motivação
A expectativa de vida é um dos índices que mostram o desenvolvimento de um país. O fato de países como Andorra (cuja expectativa de vida é de 84.5 anos) e Lesoto (no qual a expectativa é de míseros 47.1 anos) serem tão discrepantes quando se compara essa variável vai muito além da questão continental.
Com base nos conhecimentos prévios ao assunto e com a ajuda do GapMinder, resolvemos ver como a disponibilidade de alimentação de uma certa população (medida em consumo de alimentos em calorias/pessoa/dia), o saneamento básico (em porcentagem de lares com acesso ao esgoto), a mortalidade infantil (em mortes de crianças com menos de 5 anos por 1000 nascimentos) e a conclusão do ensino médio (em porcentagem da população que completou os estudos nessa fase) afeta a expectativa de vida.
Faremos então uma regressão linear múltipla que nos ajude a além de fixar os conceitos de regressão estudados em Ciência dos Dados nesse semestre, explicar como alguns fatores interferem nesse parâmetro, bastante comum para se medir o desenvolvimento de um país.

```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import pandas as pd
import numpy as np

import statsmodels.api as sm1 #Usa OLS com X como matriz de planejamento
import statsmodels.formula.api as sm  #Usa OLS como no R

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = dados = pd.read_excel('Dados.xlsx',index_col=0)
```

```python
dados=dados[~np.isnan(dados.Life)]
dados=dados[~np.isnan(dados.Child)]
dados=dados[~np.isnan(dados.Sanitation)]
dados=dados[~np.isnan(dados.School)]
dados=dados[~np.isnan(dados.Food)]
print(dados.shape)
dados.head()
```

Verificando os gráficos de dispersão e a tabela de correlações, podemos verificar como as variáveis explicativas interferem em nossa variável resposta. Resumidamente, as variáveis têm o seguinte efeito na expectativa de vida:
<ul>

<li><b>Ingestão de calorias:</b> afeta moderada e positivamente a variável resposta, ou seja, quanto maior a ingestão de calorias, maior a expectativa de vida. É a variável que menos interfere dentre as listadas.</li>
</ul>

Faremos então a regressão múltipla utilizando todas essas variáveis pelo método dos mínimos quadrados (Least Squares na tabela abaixo), com a variável Expectativa de Vida como resposta.

# Análise de Regressão

```python
#Montagem dos dados
y=dados.Life #Variável resposta: Expectativa de Vida
X=dados.iloc[:,(1,2,3,4)] #Variáveis explicativas
X=sm1.add_constant(X)  #Adiciona uma constante às variáveis explicativas

est=sm.OLS(y, X)
est=est.fit()
est.summary()
```

# Equação da Reta

$\hat{ExpVida} = 72.6337 + (-0.1663 * Child) + (0.0357 * Sanitation) + (-0.065 * School) + (0.0026 * Food) + Ei$

$\hat{y} = β_0 + (β_1 * x_1) + (β_2 * x_2) + (β_3 * x_3) + (β_4 * x_4) + Ei$

As duas equações acima representam, respectivamente, o modelo criado e a equação característica da regressão linear para o número de variáveis selecionado. Com exceção de $β_0$, que é o intercepto, os coeficientes β indicam como e em que medida as variáveis explicativas interferem na variável resposta. Se o coeficiente é negativo, a proporcionalidade entre essa variável e a resposta é inversa, logo, quanto maior for a variável desse coeficiente, menor será a resposta; quando o coeficiente é positivo, então, ocorre o contrário. Seguindo essa lógica, quando a Taxa de Mortalidade aumenta em 1, a expectativa do país diminui em 0.1663; se a % de Saneamento básico aumenta em 1%, a expectativa de vida aumenta em 0.0357; se a % de Conclusão do Ensino Médio aumenta em 1%, a ExpVida diminui 0.065 e se o consumo calórico aumenta em 1, a ExpVida aumenta em 0.026. Por fim, o intercepto é a média de expectativa de vida dos dados observados e o erro $Ei$ é o erro correspondente a cada variável explicativa.

   Para entender melhor os resultados gerados pela tabela de regressão acima, teremos de entender o significado das ferramentas, avaliando se há variáveis relevantes no modelo, a adequação das suposições e qualidade do ajuste. Para tanto, vamos começar analisando o coeficiente de determinação <b>R²</b> (R-Squared na tabela) que reflete quão bem o modelo encaixa nos dados observados. O R² varia entre 0 e 1 e quanto maior o R² do modelo, melhor ele consegue explicar os dados observados. Por obter um R² de 0,785, pode-se concluir que em 78,5% dos casos o modelo consegue explicar os dados observados.
   
   Observando o <b>R²-Ajustado</b> (Adjusted R-Squared na tabela) percebe-se que ele é menor que o R². Isso ocorre porque o R²-Ajustado leva em conta o número de variáveis explicativas (ou graus de liberdade) e o número de observações do modelo, dada por n (No. Observations na tabela), porém ele também determina quão bem o modelo se encaixa nos dados observados.
   
   Avaliando agora o parâmetro <b>teste F e probabilidade de teste F</b> (F-statistic e Prob F-Statistic na tabela), temos que ele é equivalente à vários testes t para cada variável explicativa, com o objetivo de verificar a qualidade delas. Primeiramente, o teste F, por ser um teste de hipóteses, possui uma hipótese nula e uma hipótese alternativa. A nula é a premissa de que os dados podem ser modelados precisamente quando os coeficientes das variáveis explicativas são zero. A alternativa é a premissa de que pelo menos um desses coeficientes é diferente de zero. Pelo valor da probabilidade "8.16e-38", podemos verificar que é quase impossível não rejeitar a hipótese nula, portanto, temos variáveis relevantes para o modelo.
   
   Depois, temos na tabela a <b>Probabilidade de Log</b> (Log-Likelihood na tabela), que é uma função dos parâmetros do modelo. Ela avalia a probabilidade das variáveis explicativas serem melhor explicadas com a utilização do logaritmo natural. Por termos um valor de "-341.50", conclui-se que é altamente improvável que algum dos parâmetros ser melhor explicado quando descritos com o logaritmo natural.
   
   Ainda, temos o <b>AIC</b> (Akaike information criterion) e o <b>BIC</b> (Bayesian information criterion) que são baseados na Probabilidade de Log e introduzem uma penalidade sobre a complexidade do modelo, indicando que quanto menor o valor desses coeficientes, mais simples é o modelo e menor a perda de informação entre modelos. Comparando esses valores utilizando mais ou menos variáveis explicativas, podemos perceber que não há uma grande diferença desse coeficiente, mas ganhamos mais no R², então optamos por manter esses valores.

   Passando para a próxima tabela, há a descrição mais detalhada das variáveis e pode-se perceber a coluna do <b>teste t $P>|t|$</b>, que indica a probabilidade do p-value de cada variável ser maior que o módulo do t observado para um nível de significância de 5%. Nessa coluna é preferível que os valores sejam baixos, pois quanto menor o valor desse parâmetro, menor a chance dessa variável ser irrelevante para o modelo. O teste t, assim como o teste f já mencionado, serve para verificar a qualidade das variáveis explicativas e sua relevância no modelo. No nosso caso, a variável Sanitation possui uma probabilidade significatica de $P>|t|$, porém, quando retiramos ela do modelo, nosso R² diminui um pouco, por exemplo. Assim, optamos por manter essa variável, já que ela parece estar sendo relevante.
   
   Na última tabela, temos os testes de normalidade <b>Omnibus</b> e <b>Jarque-Bera</b> que consideram <b>Skew</b> e <b>Kurtosis</b>, também representados na tabela para verificar se os resíduos do modelo têm uma distribuição normal. A diferença é que o teste Omnibus verifica se esses resíduos possuem média zero. Um valor pequeno para as probabilidades desses testes indica que a hipótese nula dificilmente será rejeitada (no nosso caso, será quase impossível), concluindo-se que os resíduos têm média zero e são normais.
   
   Ainda na última tabela, temos o teste de <b>Durbin-Waston</b> que checa a autocorrelação entre os resíduos, ou seja, checa a relação entre valores separados uns dos outros para diferentes valores de tempo. Esse parâmetro varia de 0 a 4, sendo que perto de 0 indica que há autocorrelação negativa e 4 positiva, enquanto que próximo de 2 sugere que nãoo há autocorrelação.

```python
#Linear Regression - PRIMEIRA MANEIRA DE FAZER REGRESSÃO USANDO PACOTE ols
#Escreve modelo como no R

model = sm.ols(formula='Life ~ Sanitation + Child', data = dados)
fit = model.fit()
fit.summary()

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

x_surf = np.arange(0, 100, 1)                # generate a mesh
y_surf = np.arange(0, 200, 1)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

exog = pd.core.frame.DataFrame({'Sanitation': x_surf.ravel(), 'Child': y_surf.ravel()})
out = fit.predict(exog = exog)
ax.plot_surface(x_surf, y_surf,
                out.reshape(x_surf.shape),
                rstride=1,
                cstride=1,
                color='None',
                alpha = 0.1)

ax.scatter(dados['Sanitation'], dados['Child'], dados['Life'],
           c='blue',
           marker='o',
           alpha=0.7)

ax.set_xlabel('% da população com Saneamento Básico')
ax.set_ylabel('Taxa de Mortalidade Infantil (mortes por 1000 nascimentos)')
ax.set_zlabel('Expectativa de Vida (em anos)')
```

Analisando o gráfico, podemos perceber que o modelo consegue abranger bem os pontos extremos e centrais, evidenciando que é efetivo para explicar os dados observados.

```python
fit.summary()
```

No gráfico 3D acima, utilizamos menos variáveis explicativa para conseguirmos plotá-lo, porém, utilizamos aquelas com maior correlação com nossa varíavel resposta e novamente utilizamos a tabela de regressão. Verificando todos os valores já explicados, podemos concluir que não há uma diferença tão significativa nos valores obtidos em relação ao modelo com mais variáveis, podendo ser feita essa simplificação. Contudo, é interessante que o R² seja maior, então, se conveniente, o modelo com mais variáveis é um pouco mais eficiente.

## Modelo com subgrupos

Para verificar a validade de nosso modelo de regressão, iremos avaliar se ele é igualmente bom para um certo subgrupo, ou seja, se o modelo global é igualmente válido para um suposto modelo local. Para tanto, decidimos escolher dois extremos: os 10 países com a maior expectativa de vida e os 10 com a menor.

```python
sub1 = dados[0:10]
sub2 = dados[-10:]
sub3 = [sub1,sub2]
sub4 = pd.concat(sub3)

#Montagem dos dados
y1=sub4.Life #Variável resposta: Expectativa de Vida
X1=sub4.iloc[:,(1,2,3,4)] #Variáveis explicativas
X1=sm1.add_constant(X1)  #Adiciona uma constante às variáveis explicativas

est1=sm.OLS(y1, X1)
est1=est1.fit()
est1.summary()
```

```python
model = sm.ols(formula='Life ~ Sanitation + Child', data = sub4)
fit = model.fit()
fit.summary()

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

x_surf = np.arange(0, 100, 1)                # generate a mesh
y_surf = np.arange(0, 200, 1)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

exog = pd.core.frame.DataFrame({'Sanitation': x_surf.ravel(), 'Child': y_surf.ravel()})
out = fit.predict(exog = exog)
ax.plot_surface(x_surf, y_surf,
                out.reshape(x_surf.shape),
                rstride=1,
                cstride=1,
                color='None',
                alpha = 0.1)

ax.scatter(sub4['Sanitation'], sub4['Child'], sub4['Life'],
           c='blue',
           marker='o',
           alpha=0.7)

ax.set_xlabel('% da população com Saneamento Básico')
ax.set_ylabel('Taxa de Mortalidade Infantil (mortes por 1000 nascimentos)')
ax.set_zlabel('Expectativa de Vida (em anos)')
```
